```python
# Import libraries
import os
import os.path as op
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
# Load input image
img = nib.load(op.join(os.getcwd(), 'inputs', 'ADC.nii'))
data = img.get_fdata()
print("Image shape:", data.shape)
print("Voxels size:", img.header.get_zooms())
```

```python
r_data = np.rot90(data)

# Select image slice for visualization
# Coronal
slice_cor = r_data[:][128][:]
# Axial
slice_axi = r_data[:][:][9]
```

```python
# Plot slices 
plt.figure(figsize=(15, 10))
plt.subplot(121)
plt.imshow(slice_cor, cmap='gray')
plt.title('Coronal slice')
plt.subplot(122)
plt.imshow(slice_axi, cmap='gray')
plt.title('Axial slice')
# plt.axis('off')
```
